```python
from wordcloud import WordCloud 
def createWordCloud(df, column):
    text = df[column].tolist()
    joined_text = " ".join(text)
    cloud = WordCloud(background_color="white").generate(joined_text)
    
    return cloud
```

```python
df= pd.read_csv('predicted_tweets_enhanced.csv')
```

```python
df_negativ=df[df['sentiment']=='Negativ']
```

```python
df_Postiv_neutral=df[df['sentiment']=='Postiv/neutral']
```

```python
df_Postiv_neutral.head()
```

creating a wordcloud for negativ sentiments 

```python
plt.figure(figsize=(20,10))
wordcloud=createWordCloud(df_negativ,'text')
plt.title("cloud of Negative sentiments")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

Creating a wordcloud for postiv/neutral sentiments

```python
df_Postiv_neutral['text'] = df_Postiv_neutral['text'].astype(str)
plt.figure(figsize=(20,10))
wordcloud=createWordCloud(df_Postiv_neutral,'text')
plt.title("cloud of Postiv/neutral sentiments")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

### combined Wordcloud

```python
df['text'] = df['text'].astype(str)
plt.figure(figsize=(20,10))
wordcloud=createWordCloud(df_Postiv_neutral,'text')
plt.title("Combined Wordcloud")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

Histogram to compare between the sntiments numbers

```python
import seaborn as sns
plt.figure(figsize=(10,10))
sns.countplot(x='sentiment',data=df)
plt.title("Bar chart of the sentiments")
```

```python

#define Seaborn color palette to use
plt.figure(figsize=(20,10))
colors = sns.color_palette('pastel')[0:5]
sentiments = df['sentiment'].value_counts()

#create pie chart
plt.pie(sentiments, labels=['Postive/Neutral','Negativ'] ,colors = colors, autopct='%.0f%%')
plt.title("pie chart of the sentiments")

plt.show()
```
